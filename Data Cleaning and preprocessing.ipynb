```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
data=pd.read_csv('Email Spam Detection.csv',encoding='latin-1')
data
```

```python
#Removing unwanted Columns
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
data
```

```python
#Checking for the duplicate values in the Dataframe
data.drop_duplicates(inplace=True)
print("Number of duplicate rows after cleaning:", data.duplicated().sum())
```

```python
data.describe()
```

```python
#Checking missing values in the Dataset
missing_value=data.isnull().sum()
missing_value
```

```python
#To convert values in v2 to lower case of the DataFrame
data['v2']=data['v2'].apply(lambda x: x.lower())
data
```

```python
#Function for preprocessing
def preprocess(text):
    #Removing non-alphabetical characters
    text=re.sub(r"[^a-zA-Z]", " ",text)
    
    #Tokenization and removal of stop words
    words=word_tokenize(text)
    stop_words=set(stopwords.words('english'))
    words=[word for word in words if word not in stop_words]
    return words
```

```python
#Creating a new column for tokenized words
data['tokenized_text']=data['v2'].apply(preprocess)
data.head(15)
```
