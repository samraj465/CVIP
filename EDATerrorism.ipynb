```python
data['approxdate'] = pd.to_datetime(data['approxdate'], errors='coerce')
```

```python
unique_values = data['related'].unique()
unique_values
```

```python
def handle_related_column(x):
    if isinstance(x, str):
        return ', '.join(set(x.split(', ')))
    return x

# Apply the custom function to the 'related' column
data['related'] = data['related'].apply(handle_related_column)
```

```python
# Distribution of Attack Types
plt.figure(figsize = (6,4))
sns.countplot(x='attacktype1_txt', data=data, order=data['attacktype1_txt'].value_counts().index)
plt.title('Distribution of Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distributions of regions 
plt.figure(figsize = (6,4))
sns.countplot(x='region_txt', data=data, order=data['region_txt'].value_counts().index)
plt.title('Distribution of Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of Terrorist groups
plt.figure(figsize = (6,4))
sns.countplot(x='gname', data=data, order=data['gname'].value_counts().iloc[:10].index)
plt.title('Active Terrorist Groups')
plt.xlabel('T. Groups')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count of Weapon types
plt.figure(figsize = (6,4))
sns.countplot(x = data.weaptype1_txt)
plt.title('Weapon Usage')
plt.xticks(rotation=90)
plt.show()
```

```python
# Daily attacks
plt.figure(figsize = (6,4))
sns.countplot(x = data.iday)
plt.xticks(rotation=90)
plt.title('Attacks daily')
plt.show()
```

```python
# Monthly attacks 
plt.figure(figsize = (6,4))
sns.countplot(x = data.imonth)
plt.xticks(rotation=90)
plt.title('Attacks monthly')
plt.show()
```

```python
#yearly attacks 
plt.figure(figsize = (8,6))
sns.countplot(x = data.iyear)
plt.xticks(rotation=90)
plt.title('Attacks yearly')
plt.show()
```

```python
#attack ratio by every country
plt.figure(figsize = (35,5))
sns.countplot(x = data.country)
plt.xticks(rotation=90)
plt.title('Attack ratio by country')
plt.show()
```

```python
pd.crosstab(data.iyear, data.region_txt).plot(kind = 'area', figsize = (12,6))
plt.title('Terrorism by area/region in every year')
plt.ylabel('number of attacks')
plt.show()

```

```python
#to find most attacks 
print('Most attack types: ', data['attacktype1_txt'].value_counts().idxmax())
print('City with most attacks: ', data['city'].value_counts().idxmax())
print('Country with most attacks: ', data['country'].value_counts().idxmax())
print('Year with most attacks: ', data['iyear'].value_counts().idxmax())
print('Months with most attacks: ', data['imonth'].value_counts().idxmax())
print('Region with most attacks: ', data['region_txt'].value_counts().idxmax())
```
